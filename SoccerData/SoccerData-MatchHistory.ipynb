```python
import soccerdata as sd
import pandas as pd

# Show all cols and rows
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# Create scraper class instance
mh = sd.MatchHistory(leagues="ITA-Serie A", seasons=2023)
print(mh.__doc__)
```

### Historic match results and betting odds

```python
hist = mh.read_games()
hist.head()
```

```python
#hist.reset_index(inplace=True)
hist.to_json("./data/MatchHistory/hist_match_results_and_betting_oods.json")
```
